### Line

```python
x = [1,2,3]
y = [4,5,1]

plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [9,15,7]

plt.plot(x,y,'b', label = 'Line 1')  # b for blue color
plt.plot(x2,y2,'r', label = 'Line 2', linewidth = 5)

plt.title('Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend() # To show legend as top right corner (Make sure to add labels for each line)
plt.grid(True, color = 'k')  # k for Black color


plt.show()
```

### Bar

```python
x = [5,7,9]
y = [12,16,6]

plt.bar(x,y, label = 'Bar 1')

plt.title('Bar Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend() # To show legend as top right corner (Make sure to add labels for each line)

plt.show()
```

```python
x = [1,3,5,7,9]
y = [5,2,7,8,2]

x2 = [2,4,6,8,10]
y2 = [8,6,2,5,6]

plt.bar(x,y, label = 'Bar 1', color = 'b')  # b for blue color
plt.bar(x2,y2, label = 'Bar 2', color = 'g')  

plt.title('Bar Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend() # To show legend as top right corner (Make sure to add labels for each line)

plt.show()
```

### Histogram

```python
#Width of each bin is = (max value of data  min value of data) / total number of bins 
#The default value of the number of bins to be created in a histogram is 10. However, 
#we can change the size of bins using the parameter bins in matplotlib.pyplot.hist(). 


age = np.random.randint(5,100,30)

plt.hist(age, bins = 10) 
plt.title('Hist Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

### Scatter

```python
x = [1,2,3,4,5,6,2,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label = 'Type 1')

plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend() # To show legend as top right corner (Make sure to add labels for each line)

plt.show()
```

```python
x = [1,2,3,4,5,6,2,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label = 'Type 1', color = 'navy', s = 100, marker = 'x')  # s is for Size

plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend() # To show legend as top right corner (Make sure to add labels for each line)

plt.show()
```

### Pie

```python
x = [7,2,3,13]
y = ['Sleeping', 'Eating' , 'Working' , 'Playing']

plt.pie(x, labels = y)

plt.title('Pie Chart')

plt.show()
```

```python
x = [7,2,3,13]
y = ['Sleeping', 'Eating' , 'Working' , 'Playing']

plt.pie(x, 
        labels = y, 
        colors = ['c','m','r','b'], 
        startangle = 90, 
        shadow = True,
        explode = (0,0.1,0,0),
        autopct = '%1.1f%%')  # to annote % 

plt.title('Pie Chart')
plt.legend()

plt.show()
```

### Sub Plot (Multiple Plots)

```python
plt.subplot(211)  # 211 ---> 2 means 2 Plots , 1 means 1 Horizontal Plot, 1 means 1st Graph in Veritcal position

x = [1,3,5,7,9]
y = [5,2,7,8,2]
x2 = [2,4,6,8,10]
y2 = [8,6,2,5,6]

plt.bar(x,y, label = 'Bar 1') 
plt.bar(x2,y2, label = 'Bar 2')  

plt.title('Bar Plot')


plt.subplot(212)

age = np.random.randint(5,100,30)
plt.hist(age, bins = 10)

plt.title('Hist Plot')

plt.show()
```

```python
plt.subplot(221)  # 211 ---> 2 means 2 Plots , 2 means 2 Horizontal Plot, 1 means 1st Graph

x = [1,3,5,7,9]
y = [5,2,7,8,2]
x2 = [2,4,6,8,10]
y2 = [8,6,2,5,6]

plt.bar(x,y, label = 'Bar 1') 
plt.bar(x2,y2, label = 'Bar 2')  

plt.title('Bar Plot')


plt.subplot(222)

age = np.random.randint(5,100,30)
plt.hist(age, bins = 10)

plt.title('Hist Plot')
plt.show()
```

```python
df = sns.load_dataset('flights')  # laods data from Git (https://github.com/mwaskom/seaborn-data)
df.head()
```

### Rel Plot

```python
sns.relplot(x = 'passengers', y = 'month', data = df )
plt.show()
```

```python
sns.relplot(x = 'passengers', y = 'month', hue = 'year', data = df )
plt.show()
```

### Line

```python
sns.lineplot(x = 'year', y = 'passengers', data = df)
plt.show()
```

```python
sns.lineplot(x = 'year', y = 'passengers', data = df, err_style=None)
plt.show()
```

### Distribution Plot

```python
d = np.random.normal(loc=5,size =100,scale =2)

sns.distplot(d)
plt.show()
```

```python
d = np.random.normal(loc=5,size =100,scale =2)

sns.distplot(d, hist = False)
plt.show()
```

### Joint Plot

```python
iris = sns.load_dataset('iris')

sns.jointplot(x = 'sepal_length', y = 'petal_length', data = iris)
plt.show()
```

### Box Plot

```python
sns.boxplot(x = 'month', y = 'passengers', data = df)
plt.xticks(rotation=90)  # to rotate X labels by 90
plt.show()
```

### Bar Plot

```python
sns.barplot(x = 'month', y = 'passengers', data = df)
plt.show()
```

```python
plots = sns.barplot(df['month'], df['passengers'], errwidth=0) # errwidth to hide vertical Black line for each bar
plt.xticks(rotation=90)

#To annote % for each Bar
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() /2, 
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 10),
                   textcoords='offset points')
    

plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x = 'month', y = 'passengers', data = df, hue = 'year')
plt.show()
```

### Heat Map

```python
sns.heatmap(iris.corr())
plt.show()
```
